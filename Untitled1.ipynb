```python
from pathlib import Path

MODEL_DIR = r"C:\Users\abirk\Desktop\ai_detector"

model_path = Path(MODEL_DIR)

print("Folder exists:", model_path.exists())
print("Files inside:")

for item in model_path.iterdir():
    print("-", item.name)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained(
    MODEL_DIR,
    local_files_only=True
)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_DIR,
    local_files_only=True
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)
model.eval()

print("Model loaded successfully.")
print("Using device:", device)
print("Labels:", model.config.id2label)
```

```python
import spacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")
nlp.max_length = 3000000

def analyze_with_spacy(text, max_examples=8):
    if len(text) + 100 > nlp.max_length:
        nlp.max_length = len(text) + 100

    doc = nlp(text)

    sentences = [
        sentence.text.strip()
        for sentence in doc.sents
        if sentence.text.strip()
    ]

    words = [
        token
        for token in doc
        if not token.is_space and not token.is_punct
    ]

    word_texts = [
        token.text.lower()
        for token in words
    ]

    unique_words = set(word_texts)

    pos_counts = Counter(
        token.pos_
        for token in words
    )

    important_pos = [
        "NOUN",
        "VERB",
        "ADJ",
        "ADV",
        "PROPN",
        "PRON"
    ]

    pos_examples = {}

    for pos in important_pos:
        lemmas = [
            token.lemma_.lower()
            for token in words
            if token.pos_ == pos and token.lemma_.strip()
        ]

        pos_examples[pos] = Counter(lemmas).most_common(max_examples)

    entities = [
        (entity.text, entity.label_)
        for entity in doc.ents
    ]

    entity_counts = Counter(
        label
        for text, label in entities
    )

    average_sentence_length = 0

    if len(sentences) > 0:
        average_sentence_length = len(words) / len(sentences)

    lexical_diversity = 0

    if len(words) > 0:
        lexical_diversity = len(unique_words) / len(words)

    result = {
        "sentence_count": len(sentences),
        "word_count": len(words),
        "unique_word_count": len(unique_words),
        "average_sentence_length": average_sentence_length,
        "lexical_diversity": lexical_diversity,
        "pos_counts": dict(pos_counts),
        "pos_examples": pos_examples,
        "entity_counts": dict(entity_counts),
        "entities": entities[:max_examples]
    }

    return result


def print_spacy_analysis(analysis):
    print("=" * 60)
    print("LINGUISTIC ANALYSIS WITH SPACY")
    print("=" * 60)

    print()
    print("Text statistics:")
    print("Sentences:", analysis["sentence_count"])
    print("Words:", analysis["word_count"])
    print("Unique words:", analysis["unique_word_count"])
    print(f"Average sentence length: {analysis['average_sentence_length']:.2f} words")
    print(f"Lexical diversity: {analysis['lexical_diversity']:.2f}")

    print()
    print("POS tagging counts:")

    for pos, count in analysis["pos_counts"].items():
        print(f"- {pos}: {count}")

    print()
    print("POS examples:")

    for pos, examples in analysis["pos_examples"].items():
        if len(examples) > 0:
            example_text = ", ".join(
                f"{word}={count}"
                for word, count in examples
            )

            print(f"- {pos}: {example_text}")

    print()
    print("Named entity counts:")

    if len(analysis["entity_counts"]) == 0:
        print("- No named entities found")
    else:
        for entity_type, count in analysis["entity_counts"].items():
            print(f"- {entity_type}: {count}")

    print()
    print("Named entity examples:")

    if len(analysis["entities"]) == 0:
        print("- No named entity examples")
    else:
        for entity, entity_type in analysis["entities"]:
            print(f"- {entity}: {entity_type}")
```

```python
import torch
import torch.nn.functional as F
from pprint import pprint

id2label = {
    0: "HUMAN",
    1: "AI"
}

def detect_ai_vs_human(text, top_k=10, max_tokens_to_check=80):
    linguistic_analysis = analyze_with_spacy(text)
    print_spacy_analysis(linguistic_analysis)

    model.eval()

    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=512,
        padding=True
    )

    inputs = {
        key: value.to(device)
        for key, value in inputs.items()
    }

    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]

    with torch.no_grad():
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

    probabilities = F.softmax(outputs.logits, dim=-1)[0]

    human_probability = float(probabilities[0])
    ai_probability = float(probabilities[1])

    predicted_id = int(torch.argmax(probabilities).item())
    prediction = id2label[predicted_id]
    original_confidence = float(probabilities[predicted_id])

    special_token_ids = set(tokenizer.all_special_ids)

    if tokenizer.mask_token_id is not None:
        replacement_token_id = tokenizer.mask_token_id
    else:
        replacement_token_id = tokenizer.unk_token_id

    explanation = []

    token_ids = input_ids[0].detach().cpu().tolist()

    positions_to_check = []

    for position, token_id in enumerate(token_ids):
        if token_id not in special_token_ids:
            positions_to_check.append(position)

    positions_to_check = positions_to_check[:max_tokens_to_check]

    for position in positions_to_check:
        masked_input_ids = input_ids.clone()
        masked_input_ids[0, position] = replacement_token_id

        with torch.no_grad():
            masked_outputs = model(
                input_ids=masked_input_ids,
                attention_mask=attention_mask
            )

        masked_probs = F.softmax(masked_outputs.logits, dim=-1)[0]
        masked_confidence = float(masked_probs[predicted_id])

        influence = original_confidence - masked_confidence

        token = tokenizer.convert_ids_to_tokens(
            int(input_ids[0, position].detach().cpu().item())
        )

        token = token.replace("##", "")

        explanation.append((token, influence))

    explanation = sorted(
        explanation,
        key=lambda x: x[1],
        reverse=True
    )

    explanation = explanation[:top_k]

    result = {
        "prediction": prediction,
        "human_probability": human_probability,
        "ai_probability": ai_probability,
        "explanation": explanation,
        "linguistic_analysis": linguistic_analysis
    }

    print()
    print("=" * 60)
    print("AI VS HUMAN DETECTION RESULT")
    print("=" * 60)

    print()
    print("Prediction:", prediction)
    print(f"Human probability: {human_probability * 100:.2f}%")
    print(f"AI probability:    {ai_probability * 100:.2f}%")

    print()
    print("Why did the model choose this?")
    print(f"These tokens pushed the model more toward {prediction}:")

    for token, influence in explanation:
        print(f"- {token}: influence {influence:.4f}")

    print()
    print("Reminder:")
    print("This detector is not perfect. It gives probability, not proof.")

    return result
```
