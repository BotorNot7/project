```python
import os
import numpy as np
import pandas as pd
import torch

KAGGLE_DATASET = "mishh01/ai-human-master-dataset"
MODEL_NAME = "distilbert-base-uncased"

MAX_LENGTH = 128
EPOCHS = 1
TEST_SIZE = 0.2
SEED = 24

np.random.seed(SEED)
torch.manual_seed(SEED)
```

```python
import os
import kagglehub

dataset_path = kagglehub.dataset_download(KAGGLE_DATASET)

print("Dataset downloaded to:")
print(dataset_path)

all_files = []

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        full_path = os.path.join(root, file)
        all_files.append(full_path)

print("\nFiles found:")
for file in all_files:
    print(file)
```

```python
from datasets import load_dataset

csv_files = [f for f in all_files if f.lower().endswith(".csv")]
csv_files.sort()

if len(csv_files) == 0:
    raise ValueError("No CSVs.")

csv_file = csv_files[0]

print(csv_file)

raw_dataset = load_dataset("csv", data_files={"train": csv_file})

first_split = "train"
dataset = raw_dataset[first_split]

print(raw_dataset)

print("\nColumns:")
print(dataset.column_names)

print("\nFirst row:")
print(dataset[0])
```

```python
import kagglehub
from pathlib import Path
from datasets import load_dataset

dataset_folder = Path(kagglehub.dataset_download("mishh01/ai-human-master-dataset"))

csv_files = list(dataset_folder.rglob("*.csv"))

print("CSV files found:")
for file in csv_files:
    print(file.name)

if len(csv_files) == 0:
    raise ValueError("No CSV file found.")

if len(csv_files) > 1:
    raise ValueError("More than one CSV file found. Choose one manually.")

csv_file = csv_files[0]

raw_dataset = load_dataset("csv", data_files={"train": str(csv_file)})

dataset = raw_dataset["train"]

print(dataset)

print("\nColumns:")
print(dataset.column_names)

print("\nFirst row:")
print(dataset[0])
```

```python
from collections import Counter

TEXT_COL = "text"
LABEL_COL = "label"

def clean_row(row):
    text = str(row[TEXT_COL]).strip()
    label = int(row[LABEL_COL])

    return {
        "text": text,
        "label": label
    }

standardized_dataset = dataset.map(
    clean_row,
    remove_columns=dataset.column_names
)

standardized_dataset = standardized_dataset.filter(
    lambda row: len(row["text"]) > 0
)

label_counts = Counter(standardized_dataset["label"])

print("Standardized dataset:")
print(standardized_dataset)

print("\nColumns:")
print(standardized_dataset.column_names)

print("\nFirst row:")
print(standardized_dataset[0])

print("\nLabel counts:")
print("Human:", label_counts[0])
print("AI:", label_counts[1])
```

```python
from collections import Counter
from datasets import DatasetDict

TEST_SIZE = 0.2

split_data = standardized_dataset.train_test_split(
    test_size=TEST_SIZE,
    seed=SEED
)

dataset = DatasetDict({
    "train": split_data["train"],
    "test": split_data["test"]
})

train_counts = Counter(dataset["train"]["label"])
test_counts = Counter(dataset["test"]["label"])

print(dataset)

print("\nTrain label counts:")
print("Human:", train_counts[0])
print("AI:", train_counts[1])

print("\nTest label counts:")
print("Human:", test_counts[0])
print("AI:", test_counts[1])

print("\nTrain example:")
print(dataset["train"][0])

print("\nTest example:")
print(dataset["test"][0])
```

```python
from transformers import AutoTokenizer, DataCollatorWithPadding

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

def tokenize(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        max_length=MAX_LENGTH,
        padding=False
    )

tokenized_dataset = dataset.map(
    tokenize,
    batched=True,
    batch_size=1000,
    remove_columns=["text"]
)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

print(tokenized_dataset)

print("\nFirst tokenized training row:")
print(tokenized_dataset["train"][0])
```

```python
import torch
from transformers import AutoModelForSequenceClassification

id2label = {
    0: "HUMAN",
    1: "AI"
}

label2id = {
    "HUMAN": 0,
    "AI": 1
}

if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=2,
    id2label=id2label,
    label2id=label2id
)

model.to(device)

print("Model loaded successfully.")
print("Using device:", device)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def compute_metrics(eval_pred):
    logits = eval_pred.predictions
    labels = eval_pred.label_ids

    predictions = np.argmax(logits, axis=1)

    return {
        "accuracy": accuracy_score(labels, predictions),
        "precision_ai": precision_score(labels, predictions, pos_label=1, zero_division=0),
        "recall_ai": recall_score(labels, predictions, pos_label=1, zero_division=0),
        "f1_ai": f1_score(labels, predictions, pos_label=1, zero_division=0)
    }

print("Metrics function ready.")
```

```python
import torch
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./ai_human_detector",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=EPOCHS,
    weight_decay=0.01,
    logging_steps=50,
    eval_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=1,
    load_best_model_at_end=True,
    metric_for_best_model="f1_ai",
    greater_is_better=True,
    report_to="none",
    fp16=torch.cuda.is_available()
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

print("Trainer is ready.")
```

```python
trainer.train()
```

```python
eval_results = trainer.evaluate()

print("Evaluation results:")
print(eval_results)

pred_output = trainer.predict(tokenized_dataset["test"])
predictions = np.argmax(pred_output.predictions, axis=-1)
true_labels = pred_output.label_ids

print("\nClassification report:")
print(classification_report(
    true_labels,
    predictions,
    target_names=["HUMAN", "AI"]
))

print("\nConfusion matrix:")
print(confusion_matrix(true_labels, predictions))
```
